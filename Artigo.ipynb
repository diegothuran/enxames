```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
import pandas as pd
from Util import join_strings, remove_punctuation
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data = pickle.load( open( "Database/movie_reviews/corpus/balanced_two_classes.pkl", "rb" ) )
```

```python
data = [(remove_punctuation(join_strings(d[0])), d[1]) for d in data]
```

```python
data = pd.DataFrame(data)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
my_stop_words = set(stopwords.words('english'))
tfidf_vectorizer = TfidfVectorizer(lowercase=True, stop_words=my_stop_words)
tfidf_matrix = tfidf_vectorizer.fit_transform(data[0].values)
```

```python
cos_matrix = cosine_similarity(tfidf_matrix)
df = pd.DataFrame(cos_matrix)
```

```python
df.to_csv('imdb_2_classes.csv')
```
